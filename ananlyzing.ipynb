```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
new_ver = pd.read_csv("/home/ubuntu/combos_cpp_shadow_new/Files/log_save.log")
old_ver = pd.read_csv("/home/ubuntu/combos_old/Files/log_save.log")
```

```python

plt.figure()

plt.plot(new_ver, [1] * len(new_ver), 'o', label='new_version')
plt.plot(old_ver, [2] * len(old_ver), 'o', label='old_version')

plt.xlabel('timeline')
plt.ylabel('versions')
plt.ylim(0, 3)

plt.title('Distribution of events in time')
plt.legend()

plt.show()
```

```python

plt.hist(old_ver, bins=50, alpha=0.5)
plt.xlabel('timeline')
plt.ylabel('how many events occurred')
plt.xlim(left=0, right=161000)
plt.title('Values\'histogram in the old version')
plt.show()

plt.hist(new_ver, bins=50, alpha=0.5, label=f'new_ver')
plt.title('Values\' histogram in the new version')
plt.xlabel('timeline')
plt.ylabel('how many events occurred')
plt.xlim(left=0, right=161000)
plt.show()
```
