```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
new_ver = pd.read_csv("/home/ubuntu/combos_cpp_shadow_new/Files/log_save.log")
old_ver = pd.read_csv("/home/ubuntu/combos_old/Files/log_save.log")
```

```python

plt.figure()

plt.plot(new_ver, [1] * len(new_ver), 'o', label='new_version')
plt.plot(old_ver, [2] * len(old_ver), 'o', label='old_version')

plt.xlabel('time to send message')
plt.ylabel('versions')
plt.ylim(0, 3)
plt.xlim(left=0.1, right=0.35)


plt.title('Distribution of "send" time')
plt.legend()

plt.show()
```
